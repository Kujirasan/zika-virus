```python
# Import Libraries
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from ipy_table import *
```

```python
#read result.txt (dictionary)
f = open("result.txt","r")
lines = f.readlines()
f.close()

result = []

for line in lines:
    line_split = line.split('|')
    detail = (
        line_split[0],
        line_split[1],
        line_split[2],
        line_split[3],
        float(line_split[4]),
        float(line_split[5])
    )
    result.append(detail)
```

```python
#split the virus
first_virus = result[:int(len(result)/2)]
second_virus = result[int(len(result)/2):]
```

```python
#sort the result
sorted_first = sorted(first_virus, key=lambda x: x[4], reverse=True)
sorted_second = sorted(second_virus, key=lambda x: x[4], reverse=True)
```

```python
#display first virus
T = sorted_first
header = ['Indo Virus','Foreign Virus','Location','Alignment','Score','Time (s)']
T.insert(0,header)
make_table(T)
apply_theme('basic')
```

```python
#display second virus
T = sorted_second
header = ['Indo Virus','Foreign Virus','Location','Alignment','Score','Time (s)']
T.insert(0,header)
make_table(T)
apply_theme('basic')
```
