```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# open file
seqs = {} #seqs['accid'] = str
accession_id = ""
seq = ""
with open("sequence.fasta",'r') as input_fh:
    for line in input_fh.readlines():
        if (line[0] == '>'):
            accession_id = line.split('|')[3]
        elif line == "\n":
            seqs[accession_id] = seq
            #print(seqs[accession_id])
            accession_id = ""
            seq = ""
        else: 
            seq += line.strip()
```

```python
s = [123,35,23,67,154,67,189,123,43,3,13,23,14,17,8,9,10] # -> ganti dict
sorted_index = sorted(range(len(s)),key = lambda k: s[k]) # ganti s[k].attr
sorted_s = [s[i] for i in sorted_index]
sorted_s
r = range(1,len(sorted_s)+1)
plt.scatter(r,sorted_s)
plt.show()
```
